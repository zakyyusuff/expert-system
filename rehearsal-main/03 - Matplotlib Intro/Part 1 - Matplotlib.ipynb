```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = np.sqrt(x)
y1 = np.sin(x)
y2 = np.cos(x)
```

# Inisialisasi Figure dan Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(x, y1)
ax.plot(x, y2)
```

### shortcut

```python
plt.figure(figsize=(6, 6))
plt.plot(x, y1, 'r')
plt.plot(x, y2, 'b')
plt.plot()
```

```python
plt.figure(figsize=(6, 6))
plt.plot(x, y1, 'r--')
plt.plot(x, y2, 'bv')
```

```python
plt.figure(figsize=(6, 6))
plt.plot(x, y1, 'r-,')
plt.plot(x, y2, 'bo') #bisa dilihat variasinya dengan tekan shift + tab
```

```python
plt.figure(figsize=(6, 6))
plt.plot(x, y1, 'r-,')
plt.plot(x, y2, 'bo') #bisa dilihat variasinya dengan tekan shift + tab
plt.xlim(0, 20)
plt.ylim(-4, 4)
```

```python
plt.figure(figsize=(6, 6))
plt.plot(x, y1, 'r-,', linewidth=5)
plt.plot(x, y2, 'bo') #bisa dilihat variasinya dengan tekan shift + tab
plt.xlim(0, 10)
plt.ylim(-2, 2)
```

```python
plt.figure(figsize=(6, 6))
plt.plot(x, y1, 'r-,', linewidth=5)
plt.plot(x, y2, 'bo') #bisa dilihat variasinya dengan tekan shift + tab
plt.xlim(0, 10)
plt.ylim(-2, 2)
plt.title("sin dan cos")
plt.axis('equal')
plt.xlabel("x")
plt.ylabel("y")
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.plot(x, y1, 'r-,', linewidth=5)
ax.plot(x, y2, 'bo') #bisa dilihat variasinya dengan tekan shift + tab
ax.set_xlim(0, 10)
ax.set_ylim(-2, 2)
ax.set_title("sin dan cos")
ax.set_xlabel("x")
ax.set_xlabel("y")
ax.axis('equal')
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.plot(x, y1, 'r-,', linewidth=5)
ax.plot(x, y2, 'bo') #bisa dilihat variasinya dengan tekan shift + tab
ax.set(xlim=(0, 10), ylim=(-2, 2), title="sin dan cos", xlabel=("x"), ylabel=("y"))
ax.axis('equal'); #tambahkan titik koma agar hasil angka didalam kurung pada outuput hilang
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.plot(x, y1, 'r-,', linewidth=5, label="sin")
ax.plot(x, y2, 'bo', markersize=3, label="cos") #bisa dilihat variasinya dengan tekan shift + tab
ax.set(xlim=(0, 10), ylim=(-2, 2), title="sin dan cos", xlabel=("x"), ylabel=("y"))
ax.axis('equal'); #tambahkan titik koma agar hasil angka didalam kurung pada outuput hilang
ax.legend()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()

ax.plot(x, y1, 'r-,', linewidth=5, label="sin")
ax.plot(x, y2, 'bo', markersize=3, label="cos") #bisa dilihat variasinya dengan tekan shift + tab
ax.set(xlim=(0, 10), ylim=(-2, 2), title="sin dan cos", xlabel=("x"), ylabel=("y"))
ax.axis('equal'); #tambahkan titik koma agar hasil angka didalam kurung pada outuput hilang
ax.legend(loc=8, ncol=2, shadow=True, borderpad=1) #8 adalh kode lokasi dari lower center, bisa ditulis juga langsung lower center, keterangan lanjut tekan shift +tab
```

# Scatter plot

```python
df = pd.read_csv("data/circle_data.csv", sep=";") #beri sep atau delomiter bernilai ; untuk merapikan data
df.head()
```

# TEXT DAN ANNOTATION

```python
numbers = np.random.randn(100000)
```

```python
plt.figure(figsize=(6, 6))
plt.hist(numbers, bins=80, density=1);
plt.ylim(0, 1)
plt.axvline(0, color='r')
plt.text(0, 0.8, "mean=0")
```

```python
plt.figure(figsize=(6, 6))
plt.hist(numbers, bins=80, density=1);
plt.ylim(0, 1)
plt.axvline(0, color='r')
plt.text(0, 0.8, "mean=0", ha='center')
```

```python
plt.figure(figsize=(6, 6))
plt.hist(numbers, bins=80, density=1);
plt.ylim(0, 1)
plt.axvline(0, color='r')
plt.text(0, 0.6, "mean=0", ha='right')
```

```python
plt.figure(figsize=(6, 6))
plt.hist(numbers, bins=80, density=1);
plt.ylim(0, 1)
plt.axvline(0, color='r')
plt.text(0, 0.8, "mean=0", ha='center')
plt.annotate("mean=0", xy=(0, 0.4),xytext=(2, 0.5), ha='center',arrowprops={'arrowstyle' : '-|>'})
```
