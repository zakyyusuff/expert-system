# Seaborn & Pandas Dataframe

### Statistical Summary

```python
df = pd.read_csv("data/cereal.csv", index_col="name")
df.head()
```

### Missing Value

```python
#Untuk visualisasi data yang miss value
plt.figure(figsize=(15, 15))
sns.heatmap(df.isna(), cmap='Blues')
```

```python
#Untuk visualisasi data yang miss value
plt.figure(figsize=(15, 15))
sns.heatmap(df.isna(), cmap='Blues', cbar=False)
```

```python
#Untuk visualisasi data yang miss value
plt.figure(figsize=(15, 15))
sns.heatmap(~df.isna(), cmap='Blues', cbar=False) #data yg kosong adalah data yg bolong
```

### Correlation Heatmap

```python
#Untuk visualisasi data korelasi, korelasi adalah data yang saling mempengaruhi
plt.figure(figsize=(15, 15))
sns.heatmap(df.corr(), cmap='bwr', cbar=False)
```

```python
#Untuk visualisasi data yang miss value
plt.figure(figsize=(15, 15))
sns.heatmap(df.corr(), cmap='bwr', cbar=False, vmin=-1, vmax=1, annot=True, square=True)
```

```python
#Untuk visualisasi data yang miss value
plt.figure(figsize=(15, 15))
sns.heatmap(df.corr(), cmap='bwr', cbar=False, vmin=-1, vmax=1, annot=True, square=True, fmt=".1f") #fmt .1f adalah 1 angka dibelakang koma
```

# pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="shelf", vars=["calories", "fat", "rating"]) #melihat pengaruh calori dan lemak terhadap rating
```
